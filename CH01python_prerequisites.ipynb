## 1.1 推导式

```python
given_list = [1, 2, 3, 4, 5]
given_list
```

```python
def my_func(x):
    return x ** 2
```

```python
new_list = []
for i in range(5):
    new_list.append(my_func(given_list[i]))
new_list
```

```python
list(my_func(it) for it in given_list)
```

```python
[my_func(it) for it in given_list]
```

```python
{my_func(it) for it in given_list}
```

```python
{it: my_func(it) for it in given_list}
```

```python
# 给定一个包含 5 个英语单词的列表，请构造一个以列表的元素为键，以每个键对应的单词字母个数为值的字典
words = ['apple', 'banana', 'cherry', 'date', 'elderberry']
word_dict = {word: len(word) for word in words}
word_dict
```

```python
sum([my_func(it) for it in given_list])
```

```python
all([my_func(it) for it in given_list])
```

```python
max([my_func(it) for it in given_list])
```

```python
[(color, item) for color in ["red", "blue"] for item in ["pen", "book"]]
```

```python
[[(color, item) for color in ["red","blue"]] for item in ["pen", "book"]]
```

## 1.2 匿名函数

```python
single = lambda x: x ** 2
single(3)
```

```python
multi = lambda a, b: a + b
multi(2, 5)
```

```python
my_list = [[1, 2], [3, 4, 5], [6], [7, 8], [9]]
list(filter(lambda x: len(x) > 2, my_list))
```

```python
list(filter(lambda x: sum(i if i % 2 == 1 else 0 for i in x) > 5, my_list))
```
